**TITLE OF THE PROJECT:  Servo Prediction Using Linear Regression**



**Objective: The servo prediction is a control system that accurately positions or controls the position of a mechanical device, such as motor or actuator.**

Data Source: The attributes in this set are: Motor, Screw, Pgain, Vgain, Class







**Import Library:**

```python
import pandas as pd
```

**Import Data:**

```python


df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Servo%20Mechanism.csv')
```

```python
df.head()
```

```python
df.info()
```

**Describe Data:**

```python
df.describe()
```

```python
df.columns
```

```python
df.shape
```

**Data Visualization:**

```python
df[['Motor']].value_counts()
```

```python
df[['Screw']].value_counts()
```

**Data Preprocessing:**

```python
df.replace({'Motor':{'A':0, 'B':1, 'C':2,  'D' :3, 'E': 4}},inplace=True)
```

```python
df.replace({'Screw':{'A':0, 'B':1, 'C':2, 'D': 3, 'E': 4}},inplace=True)
```

**Define Target Variable(y) and Feature Variable(x):**

```python
y = df['Class']
```

```python
y.shape
```

```python
y
```

```python
x = df[['Motor', 'Screw', 'Pgain', 'Vgain']]
```

```python
x = df.drop('Class', axis=1)
```

```python
x.shape
```

```python
x
```

**Train Test Split:**


```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

**Modeling:**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

**Prediction:**

```python
y_pred = lr.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

**Model Evaluation:**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicated")
plt.show()
```
